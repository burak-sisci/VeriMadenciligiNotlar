**AdÄ±m 2: Verileri toplayÄ±n**

    <li>Ã–rneklem bÃ¼yÃ¼klÃ¼ÄŸÃ¼: n = 40 Ã¶ÄŸrenci</li>
    <li>Ã–rneklem ortalamasÄ±: $\bar{x}$  = 525</li>
    <li>NÃ¼fus standart sapmasÄ±: Ïƒ = 100 (tarihsel verilerden bilinmektedir)</li>

**AdÄ±m 3: Test istatistiÄŸini hesaplayÄ±n**
z = $\frac{\bar{x} - \mu_0}{\sigma / \sqrt{n}}$ </br>
z = (525 - 500) / (100 / âˆš40)</br>

z = 25 / 15.81</br>

**AdÄ±m 4: Z puanÄ±nÄ± yorumlayÄ±n**
1,58'lik bir z-puanÄ± ÅŸu anlama gelir: "Ã–rneklem ortalamasÄ±, Hâ‚€ doÄŸru olsaydÄ± beklediÄŸimizden 1,58 standart hata daha yÃ¼ksektir."</br>

**z = 1.58 bize ne anlatÄ±yor?**

Hâ‚€ doÄŸru olsaydÄ± (hazÄ±rlÄ±k kursu iÅŸe yaramasaydÄ±), tÃ¼m olasÄ± Ã¶rneklem ortalamalarÄ±nÄ±n daÄŸÄ±lÄ±mÄ±nÄ±n bu olduÄŸunu varsayalÄ±m. Ã–rneklem ortalamalarÄ±nÄ±n Ã§oÄŸu 500 civarÄ±nda kÃ¼melenirdi, ancak bazÄ±larÄ± ÅŸans eseri daha yÃ¼ksek veya daha dÃ¼ÅŸÃ¼k olurdu. 1,58'lik bir z puanÄ±, Ã¶rneklem ortalamamÄ±zÄ±n (525) bu daÄŸÄ±lÄ±mÄ±n Ã¼st kÄ±smÄ±nda olduÄŸunu, ancak Ã§ok da uzakta olmadÄ±ÄŸÄ±nÄ± gÃ¶sterir.

**AdÄ±m 5: Karar verin**

z = 1,58'in Hâ‚€'yi reddetmek iÃ§in "yeterince aÅŸÄ±rÄ±" olup olmadÄ±ÄŸÄ±na karar vermek iÃ§in kritik deÄŸere veya p-deÄŸerine bakÄ±yoruz (bunu bir sonraki derste ayrÄ±ntÄ±lÄ± olarak ele alacaÄŸÄ±z). </br>

**SonuÃ§:** "Bu Ã¶rneÄŸe dayanarak, hazÄ±rlÄ±k kursunun sÄ±nav puanlarÄ±nÄ± iyileÅŸtirdiÄŸi sonucuna varmak iÃ§in yeterli gÃ¼Ã§lÃ¼ kanÄ±t yok. GÃ¶zlemlediÄŸimiz 25 puanlÄ±k artÄ±ÅŸÄ±n tesadÃ¼fen gerÃ§ekleÅŸmesi makul olabilir." </br>

````python
#from statsmodels.stats.weightstats import ztest
# veriler varsa 
from math import sqrt
from scipy.stats import norm

n = 40
x_bar = 525
sigma = 100
mu0 = 500

# Z istatistiÄŸi
z = (x_bar - mu0) / (sigma / sqrt(n))

# Tek kuyruklu p-deÄŸerleri
# H1: ortalama > mu0 ise (saÄŸ kuyruk)
p_right = 1 - norm.cdf(z)

# H1: ortalama < mu0 ise (sol kuyruk)
p_left = norm.cdf(z)

# p-deÄŸeri (2 kuyruklu test)
p_twotail = 2 * (1 - norm.cdf(abs(z)))

print(f"Z istatistiÄŸi: {z:.3f}")
print(f"SaÄŸ kuyruk p-deÄŸeri (H1: >): {p_right:.5f}")
print(f"Sol kuyruk p-deÄŸeri (H1: <): {p_left:.5f}")
print(f"Ä°ki kuyruklu p-deÄŸeri: {p_twotail:.5f}")
````

**2. Two-Sample Z-Test**

**CevapladÄ±ÄŸÄ± soru:** "Ä°ki farklÄ± popÃ¼lasyonun ortalamalarÄ± anlamlÄ± derecede farklÄ± mÄ±dÄ±r?"</br>

**FormÃ¼l**

Ïƒâ‚, Ïƒâ‚‚ = popÃ¼lasyon standart sapmalarÄ± (bilinen) </br>

**Ã–rnek**

**Senaryo:** Erkek ve kadÄ±n yazÄ±lÄ±m mÃ¼hendislerinin ortalama maaÅŸlarÄ±nÄ± karÅŸÄ±laÅŸtÄ±rmak istiyorsunuz. SektÃ¶r genelindeki verilerden ÅŸunlarÄ± biliyorsunuz:</br>

55 kadÄ±n mÃ¼hendis: $\bar{x}_2$ = 820000, nâ‚‚ = 55 </br>

Hipotezler:

Hâ‚€: Î¼â‚ = Î¼â‚‚ (MaaÅŸ farkÄ± yok) </br>

Bu iki yÃ¶nlÃ¼ bir testtir Ã§Ã¼nkÃ¼ herhangi bir fark olup olmadÄ±ÄŸÄ±nÄ± (her iki yÃ¶nde de) kontrol ediyoruz. </br>

z'yi hesaplayalÄ±m:

z = (850000 - 820000) / âˆš(50000Â²/60 + 45000Â²/55)</br>

3,39'luk bir z puanÄ± oldukÃ§a uÃ§ bir deÄŸerdir! Î± = 0,05'te iki taraflÄ± bir test iÃ§in kritik deÄŸerler Â±1,96'dÄ±r. </br>

````python
from math import sqrt
from scipy.stats import norm

# Veriler
x1_bar = 850000     # 1. Ã¶rneklem ortalamasÄ±
x2_bar = 820000     # 2. Ã¶rneklem ortalamasÄ±
sigma1 = 50000      # 1. grubun popÃ¼lasyon std sapmasÄ±
sigma2 = 45000      # 2. grubun popÃ¼lasyon std sapmasÄ±
n1 = 60
n2 = 55

# Z istatistiÄŸi
z = (x1_bar - x2_bar) / sqrt((sigma1**2 / n1) + (sigma2**2 / n2))

# p-deÄŸeri (Ã§ift kuyruklu Ã¶rnek)
p = 2 * (1 - norm.cdf(abs(z)))

print(f"Z istatistiÄŸi: {z:.3f}")
print(f"p-deÄŸeri: {p:.5f}")
````

        z hesaplanÄ±r.</br>
        z = (0.85 - 0.90) / âˆš(0.90 Ã— 0.10 / 200)</br>
        z = -0.05 / âˆš(0.09 / 200)</br>
        z = -0.05 / âˆš0.00045</br>
        z = -0.05 / 0.0212</br>
        z = -2.36
    </li>
    <li>
        AdÄ±m 4: Karar Î± = 0,05'te sol kuyruklu bir test iÃ§in kritik deÄŸer -1,645'tir.
    </li>
</ul

-2.36 < -1.645 (daha uÃ§) olduÄŸundan Hâ‚€'yi reddediyoruz.</br>

SonuÃ§: "GerÃ§ek memnuniyet oranÄ± muhtemelen %90'Ä±n altÄ±ndadÄ±r. Åirketin iddiasÄ± abartÄ±lÄ± gÃ¶rÃ¼nÃ¼yor."

````python
from math import sqrt
from scipy.stats import norm

# Verilen deÄŸerler
n = 200                  # Ã¶rneklem bÃ¼yÃ¼klÃ¼ÄŸÃ¼
x = 170                  # memnun mÃ¼ÅŸteri sayÄ±sÄ±
p_hat = x / n            # Ã¶rneklem oranÄ±
p0 = 0.90                # varsayÄ±lan (hipotez edilen) oran

# Z istatistiÄŸi hesaplama
z = (p_hat - p0) / sqrt((p0 * (1 - p0)) / n)

# p-deÄŸeri (tek kuyruklu, H1: p < p0)
p_value = norm.cdf(z)

# SonuÃ§larÄ± yazdÄ±r
print(f"Z istatistiÄŸi: {z:.3f}")
print(f"p-deÄŸeri: {p_value:.5f}")

# Karar
alpha = 0.05
if p_value < alpha:
    print("H0 reddedilir: MÃ¼ÅŸteri memnuniyet oranÄ± %90'dan dÃ¼ÅŸÃ¼ktÃ¼r.")
else:
    print("H0 reddedilemez: Veriler %90 memnuniyet oranÄ±na aykÄ±rÄ± deÄŸildir.")
````

````python
from math import sqrt
from scipy.stats import t

# Veriler
x1, s1, n1 = 78, 8, 30
x2, s2, n2 = 72, 10, 28

# Pooled variance hesapla (eÅŸ varyans varsayÄ±mÄ±)
sp_squared = ((n1 - 1)*s1**2 + (n2 - 1)*s2**2) / (n1 + n2 - 2)

# t istatistiÄŸi hesapla
t_stat = (x1 - x2) / sqrt(sp_squared * (1/n1 + 1/n2))

# Serbestlik derecesi
df = n1 + n2 - 2

# Ã‡ift kuyruklu p-deÄŸeri
p_value = 2 * (1 - t.cdf(abs(t_stat), df))

# SonuÃ§larÄ± yazdÄ±r
print(f"t istatistiÄŸi: {t_stat:.3f}")
print(f"p-deÄŸeri (Ã§ift kuyruklu): {p_value:.5f}")

# Karar
alpha = 0.05
if p_value < alpha:
    print("H0 reddedilir: YÃ¶ntemler arasÄ±nda anlamlÄ± bir fark vardÄ±r.")
else:
    print("H0 reddedilemez: YÃ¶ntemler arasÄ±nda anlamlÄ± fark yoktur.")
````

**YaygÄ±n KarÄ±ÅŸÄ±klÄ±k #4: "Verilerim eÅŸleÅŸtirilmiÅŸ, ancak yine de normalliÄŸi kontrol etmeli miyim?"**
EVET! EÅŸleÅŸtirilmiÅŸ t-testi iÃ§in, farklarÄ±n yaklaÅŸÄ±k olarak normal daÄŸÄ±lÄ±m gÃ¶stermesi gerekir (veya yeterince bÃ¼yÃ¼k bir Ã¶rneÄŸiniz olmalÄ±dÄ±r).</br>

````python
from math import sqrt
from scipy.stats import t

# Verilen bilgiler
d_bar = 6.5   # farklarÄ±n ortalamasÄ±
sd = 4.2      # farklarÄ±n standart sapmasÄ±
n = 20        # hasta sayÄ±sÄ±

# t istatistiÄŸi
t_stat = (d_bar - 0) / (sd / sqrt(n))

# Serbestlik derecesi
df = n - 1

# Tek kuyruklu p-deÄŸeri (H1: d > 0)
p_value = 1 - t.cdf(t_stat, df)

# SonuÃ§larÄ± yazdÄ±r
print(f"t istatistiÄŸi: {t_stat:.3f}")
print(f"p-deÄŸeri (tek kuyruklu): {p_value:.5f}")

# Karar
alpha = 0.05
if p_value < alpha:
    print("H0 reddedilir: Ä°laÃ§ kan basÄ±ncÄ±nÄ± anlamlÄ± derecede dÃ¼ÅŸÃ¼rmektedir.")
else:
    print("H0 reddedilemez: Ä°lacÄ±n tansiyonu dÃ¼ÅŸÃ¼rdÃ¼ÄŸÃ¼ne dair yeterli kanÄ±t yoktur.")
````

````python
import numpy as np
from scipy.stats import ttest_rel

# Ã–rnek veri oluÅŸturma (simÃ¼lasyon)
# "Ã–nce" tansiyon ortalamasÄ± yaklaÅŸÄ±k 140, "Sonra" ortalamasÄ± yaklaÅŸÄ±k 133.5
np.random.seed(42)
before = np.random.normal(loc=140, scale=8, size=20)
after = before - np.random.normal(loc=6.5, scale=4.2, size=20)

# Paired t-test
t_stat, p_two_tailed = ttest_rel(before, after)

# Tek kuyruklu p-deÄŸeri (H1: before > after yani ilaÃ§ -> dÃ¼ÅŸÃ¼rÃ¼r)
if t_stat > 0:
    p_one_tailed = p_two_tailed / 2
else:
    p_one_tailed = 1 - (p_two_tailed / 2)

# SonuÃ§larÄ± yazdÄ±r
print(f"Ã–nce ortalama: {np.mean(before):.2f}")
print(f"Sonra ortalama: {np.mean(after):.2f}")
print(f"t istatistiÄŸi: {t_stat:.3f}")
print(f"p-deÄŸeri (tek kuyruklu): {p_one_tailed:.7f}")

# Karar
alpha = 0.05
if p_one_tailed < alpha:
    print("H0 reddedilir: Ä°laÃ§ kan basÄ±ncÄ±nÄ± anlamlÄ± derecede dÃ¼ÅŸÃ¼rmektedir.")
else:
    print("H0 reddedilemez: Ä°lacÄ±n tansiyonu dÃ¼ÅŸÃ¼rdÃ¼ÄŸÃ¼ne dair yeterli kanÄ±t yoktur.")
````

````python
import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency

# Ham veri oluÅŸturma
data = []

data += [["Smoker", "Disease"]] * 60
data += [["Smoker", "No Disease"]] * 140
data += [["Non-Smoker", "Disease"]] * 30
data += [["Non-Smoker", "No Disease"]] * 270

df = pd.DataFrame(data, columns=["Smoking", "LungDisease"])

# Kontenjans tablosu oluÅŸtur
contingency_table = pd.crosstab(df["Smoking"], df["LungDisease"])
print("GÃ¶zlenen Kontenjans Tablosu:")
print(contingency_table)

# Ki-Kare BaÄŸÄ±msÄ±zlÄ±k Testi
chi2_stat, p_value, dof, expected = chi2_contingency(contingency_table)

print("\nChi-Square istatistiÄŸi:", round(chi2_stat, 3))
print("Serbestlik derecesi:", dof)
print("p-deÄŸeri:", round(p_value, 7))

print("\nBeklenen DeÄŸerler:")
print(pd.DataFrame(expected,
                   index=contingency_table.index,
                   columns=contingency_table.columns))

# Karar
alpha = 0.05
if p_value < alpha:
    print("\nH0 reddedilir: Sigara iÃ§me durumu ile akciÄŸer hastalÄ±ÄŸÄ± arasÄ±nda anlamlÄ± iliÅŸki vardÄ±r.")
else:
    print("\nH0 reddedilemez: Sigara iÃ§me ile akciÄŸer hastalÄ±ÄŸÄ± arasÄ±nda iliÅŸki yoktur.")
````

````python
import numpy as np
import pandas as pd
from scipy.stats import chi2

# GÃ¶zlenen frekans tablosu
observed = np.array([[60, 140],
                     [30, 270]])

# SatÄ±r ve sÃ¼tun toplamlarÄ±
row_totals = observed.sum(axis=1)
col_totals = observed.sum(axis=0)
grand_total = observed.sum()

# Beklenen deÄŸerleri hesapla
expected = np.outer(row_totals, col_totals) / grand_total

# Ï‡Â² = Î£ (O - E)^2 / E
chi_square = np.sum((observed - expected)**2 / expected)

# Serbestlik derecesi = (satÄ±r-1)*(sÃ¼tun-1)
df = (observed.shape[0] - 1) * (observed.shape[1] - 1)

# p-deÄŸeri
p_value = 1 - chi2.cdf(chi_square, df)

# SonuÃ§larÄ± yazdÄ±r
print("GÃ¶zlenen Tablo:")
print(pd.DataFrame(observed, index=["Smoker", "Non-Smoker"], columns=["Disease", "No Disease"]))

print("\nBeklenen Tablo:")
print(pd.DataFrame(np.round(expected, 2), index=["Smoker", "Non-Smoker"], columns=["Disease", "No Disease"]))

print(f"\nChi-Square Ä°statistiÄŸi: {chi_square:.3f}")
print(f"Serbestlik Derecesi: {df}")
print(f"p-deÄŸeri: {p_value:.7f}")

# Karar
alpha = 0.05
if p_value < alpha:
    print("\nH0 reddedilir: Sigara iÃ§me durumu ile akciÄŸer hastalÄ±ÄŸÄ± arasÄ±nda iliÅŸki vardÄ±r.")
else:
    print("\nH0 reddedilemez: Sigara iÃ§me durumu ile akciÄŸer hastalÄ±ÄŸÄ± arasÄ±nda iliÅŸki yoktur.")
````
