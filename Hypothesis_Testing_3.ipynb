**Dedektif Hikayesi: İstatistiksel Kanıtları Anlamak**

Diyelim ki bir madeni paranın adil mi yoksa taraflı mı olduğunu araştıran bir dedektifsiniz.</br>

**Setup:** Biri size bir madeni para uzatıyor ve "Bu adil bir madeni para - %50 yazı, %50 tura." diyor. Ama siz şüpheleniyorsunuz. Bu yüzden denemeye karar veriyorsunuz.</br>

**Test:** Madeni parayı 20 kere havaya atıyorsunuz ve 15 kez yazı 5 kez tura geliyor.</br>

**Sorunuz:** "Eğer bu para gerçekten adil olsaydı, 20 atıştan 15'inin yazı gelme olasılığı sadece şans eseri ne kadar olurdu?"</br>

**Bu soru - bu TEK soru -** TÜM hipotez testlerinin kalbidir. Her kavramı bu bakış açısıyla incelenmelidir.</br>





**P Değerini Anlamak: İstatistikte En Çok Yanlış Anlaşılan Kavram**

**İnsanların P-Değerinin Anlamını Düşünmesi (YANLIŞ!)**

❌ "H₀'nin doğru olma olasılığı" ❌ "Sonucumun şans eseri ortaya çıkma olasılığı" ❌ "Hata yapma olasılığı" ❌ "H₁'nin doğru olma olasılığı</br>

BUNLARIN HEPSİ YANLIŞ! İstatistiklerdeki 1 numaralı kafa karışıklığı kaynağıdır.</br>

**P-Değeri Aslında Ne Anlama Geliyor (DOĞRU!)**

p-değeri:</br>

**"Eğer H₀ doğruysa, gerçekte gözlemlediğim kadar uç bir veri elde etme olasılığı nedir?"**</br>

Bunu madeni para örneğiyle dikkatlice açıklayalım.</br>

**Madeni Para Örneği: P-Değerinin Adım Adım Hesaplanması**</br>

**Hipotezlerimiz:**<br>

<ul>
    <li>H₀: Madeni para adildir (yazı için p = 0,5)</li>
    <li>Hₐ: Madeni para yazıya doğru eğilimlidir (p > 0,5)</li>
</ul>


**Gözlemimiz:** 20 atıştan 15'i yazı.</br>

**P-Değeri Sorusu:** "Eğer madeni para adil ise (H₀ doğruysa), 20 atıştan 15 veya daha fazlasının yazı gelme olasılığı nedir?"

**Bunu bir düşünelim:**

Eğer para gerçekten adilse, çoğu zaman 10 civarı yazı (20'nin yarısı) gelir. Bazen 9 veya 11, bazen 8 veya 12, nadiren 7 veya 13 ve çok nadiren 15 gibi uç bir sayı gelir.</br>

Binom dağılımını kullanarak (hesaplama detaylarıyla ilgilenmeden) şunlar bulunabilir:</br>
<ul>
    <li>P(exactly 15 heads) = 0.0148</li>
    <li>P(exactly 16 heads) = 0.0046</li>
    <li>P(exactly 17 heads) = 0.0011</li>
    <li>P(exactly 18 heads) = 0.0002</li>
    <li>P(exactly 19 heads) = 0.00002</li>
    <li>P(exactly 20 heads) = 0.000001</li>
</ul>

**p değeri = tüm bu olasılıkların toplamı = 0,0207 (yaklaşık %2)**


**Bu P-Değerinin Yorumlanması**

**p = 0,0207'nin anlamı:** "Madeni para gerçekten adil olsaydı, 20 atışta 15 veya daha fazla yazı gelme olasılığı yalnızca %2 olurdu."</br>

**Bunun bize gösterdiği şey şu:** Adil bir para için 15 kez tura gelmesi oldukça sıra dışı bir durum. İmkansız değil, ama paranın taraflı olduğundan şüphelenmemize yetecek kadar sıra dışı.</br>

**Bunun bize söylemediği şey:**
<ul>
    <li>❌ Bize madeni paranın taraflı olma olasılığını söylemiyor (bu farklı bir soru!)</li>
    <li>❌ Bize madeni paranın ne kadar taraflı olduğunu SÖYLEMEZ</li>
    <li>❌ Bize %2 olasılıkla bir hata yaptığımızı söylemiyor</li>
</ul>

**Resmi Tanım**

p değeri:</br>

P (elde ettiğimiz kadar uç gözlem verileri | H₀ doğrudur)</br>

"|" sembolü "verildiği takdirde" veya "varsayıldığı takdirde" anlamına gelir.</br>

**Önemli Bilgi:** p değeri, H₀'nin doğru olduğu varsayılarak hesaplanır. H₀'nin doğru olma olasılığı değildir!</br>

**Şöyle düşünün: H₀'yi yargılıyorsunuz.** p-değeri, H₀'nin masum olması durumunda kanıtınızın ne kadar şaşırtıcı olacağını ölçer. Küçük p-değeri = çok şaşırtıcı = H₀ suçlu görünüyor.

**Alfa'yı (α) Anlamak: Önem Düzeyi**

Artık p-değeri anlaşıldığına göre, şimdi α (alfa) hakkında konuşalım.</br>

**Alfa Nedir?**</br>
Alfa (α), H₀'yi reddetmeden önce p değerinin ne kadar küçük olması gerektiğine dair **önceden belirlenen eşiktir.**</br>

Yaygın tercihler:</br>

<ul>
    <li>α = 0,05 (%5) - en yaygın</li>
    <li>α = 0,01 (%1) - daha risksiz</li>
    <li>α = 0,10 (%10) - daha liberal</li>
</ul>

**Karar Kuralı**

**p-değeri ≤ α ise:** H₀'yi reddedin (kanıt yeterince güçlüdür) **p-değeri > α ise:** H₀'yi reddetmeyin (kanıt yeterince güçlü değildir)</br>

**Madeni Para Örneğimize Devam Edelim**

p değeri 0,0207 (2,07%) olarak hesaplandı.</br>

**α = 0,05 olarak belirlenmişse:**</br>

<ul>
    <li>p-değeri (0.0207) < α (0.05)</li>
    <li>Karar: H₀'yi Reddet</li>
    <li>Sonuç: "Madeni paranın yazı gelme eğiliminde olduğu sonucuna varmak için yeterli kanıt var"</li>
</ul>

**α = 0,01 olarak belirlenmişse:**</br>

<ul>
    <li>p-değeri (0.0207) > α (0.01)</li>
    <li>Karar: H₀ reddedilemedi</li>
    <li>Sonuç: "Madeni paranın taraflı olduğu sonucuna varmak için yeterli kanıt yok"</li>
</ul>

**Dikkat: Aynı veriler, aynı p değeri, α'ya bağlı olarak farklı sonuçlar!**


**Neden α = 0,05?**

İşte bir sır: **Bu biraz keyfi!**

α = 0,05 kullanma geleneği 1920'lerde Ronald Fisher'dan geliyor. %5'in makul bir eşik gibi göründüğünü söylemiş. Hepsi bu. Yani, %5'te büyülü bir şey yok.</br>

**α'nın Gerçek Anlamı:**

Alfa, yanlış alarmlara (Tip I hataları - bu konu hakkında daha fazla bilgi bir sonraki derste anlatılacaktır) karşı toleransınızdır.</br>

<ul>
    <li> α = 0,05 şu anlama gelir: "Yanılma olasılığım %5 olduğunda H₀'yi (bir şeyin doğru olduğunu iddia etmeyi) reddetmeye hazırım"</li>
    <li>α = 0,01 şu anlama gelir: "Daha temkinliyim - sadece %1 yanılma olasılığı istiyorum"</li>
    <li> α = 0,10 şu anlama gelir: "Daha hoşgörülüyüm - %10 yanılma olasılığını kabul ederim"</li>
</ul>

**Alfa Seçimi"</br>

**Aşağıdaki durumlarda α = 0,01 (daha katı) değerini kullanın:**</br>

<ul>
    <li>Yanlış pozitifler çok maliyetliyse</li>
    <li>Güçlü kanıtlara ihtiyaç duyan iddialı bir iddiada bulunuluyorsa</li>
    <li>Örnek: Bir ilacın kanseri iyileştirdiğini iddia etmek - çok emin olunması gerekir!</li>
</ul>


**Aşağıdaki durumlarda α = 0,05 (standart) değerini kullanın:**</br>

<ul>
    <li>Orta düzeyde sonuçlar</li>
    <li>Akademik araştırmaların çoğu</li>
</ul>

**Aşağıdaki durumlarda α = 0,10 (daha esnek) değerini kullanın:**</br>

<ul>
    <li>Keşifsel araştırma</li>
    <li>Ön çalışmalar</li>
    <li> Gerçek bir etkiyi gözden kaçırmak, yanlış pozitif bir sonuçtan daha maliyetlidir</li>
</ul>


**Kritik Bir Nokta: Veri Toplamadan ÖNCE Alfa'yı Ayarlayın**

**Kötü uygulama (p-hacking):**

<ol>
    <li>Veri topla</li>
    <li>p-değerini hesapla</li>
    <li>Sonucu anlamlı kılmak için α'yı seç</li>
    <li>"Bak, p = 0,06... yayınlayabilmek için α = 0,10 kullanalım!"</li>
</ol>

**İyi uygulama :**

<ol>
    <li>Alanınızın standartlarına ve sonuçlarına göre α'yı seçin</li>
    <li>Veri toplayın</li>
    <li>P-değerini hesaplayın</li>
    <li>"Önceden belirlediğiniz α ile karşılaştırın</li>
    <li>Kararınızı verin</li>
</ol>

**Test İstatistikleri ve Kritik Değerler: Z-puanlarını, T-puanlarını vb. Anlamak**

Şimdi bu test istatistiklerinin (z, t, χ², F) aslında neyi temsil ettiğini ve p-değerleri ve α ile nasıl ilişkili olduğunu anlamaya çalışalım.</br>

**Test İstatistiği Nedir?**</br>

Bir test istatistiği, örnek verilerinizin H₀ altında beklediğinizden ne kadar uzakta olduğunu özetleyen tek bir sayıdır.</br>

**Bunu şöyle düşünün:** "Sonucum beklenenden kaç 'standart birim' uzakta?"</br>

**Z-Puanını Derinlemesine Anlamak**

**Senaryo:** IQ testleri şu şekilde tasarlanmıştır:</br>

<ul>
    <li>Ortalama (μ) = 100</li>
    <li>Standart sapma (σ) = 15</li>
</ul>

Bir kişiyi test ediyorsunuz ve 130 puan alıyor. Bu ne kadar sıra dışı?</br>

**Z-puanının hesaplanması:**

z = (X - μ) / σ</br>
z = (130 - 100) / 15</br>
z = 30 / 15</br>
z=2</br>

**z = 2'nin anlamı:** "Bu kişinin IQ'su ortalamanın 2 standart sapma üstündedir."</br>

**Z Dağılımı (Standart Normal Dağılım)**

Z dağılımı şu şekilde bir çan eğrisidir:</br>

<ul>
    <li>Ortalama = 0</li>
    <li>Standart sapma = 1</li>
</ul>

Her z-puanı bu standart eğri üzerinde nerede olduğunuzu söyler.</br>

<ul>
    <li>Verilerin yaklaşık %68'i z = -1 ile +1 arasında değişmektedir.</li>
    <li>Verilerin yaklaşık %95'i z = -2 ile +2 arasında değişmektedir.</li>
    <li>Verilerin yaklaşık %99,7'si z = -3 ile +3 arasında değişmektedir.</li>
</ul>

Yani z = 2 ise, tipik %95 aralığının sınırındasınız. İnsanların yalnızca yaklaşık %2,5'i bu kadar yüksek veya daha yüksek puan alıyor.</br>

**Z-Puanından P-Değerine**</br>

**Sağ kuyruklu bir test için (Hₐ: μ > μ₀):**</br>

<ul>
    <li>Z-puanını hesapla</li>
    <li>p-değeri = z'yi bu kadar büyük veya daha büyük alma olasılığı</li>
    <li>Z-tablosuna bakın veya yazılım kullanın</li>
</ul>


Örnek: Sağ kuyruklu bir testte z = 2.0 ise:</br>

<ul>
    <li>p-value ≈ 0.0228 (2.28%)</li>
    <li>Bu, z = 2'nin sağındaki eğrinin altındaki alandır</li>
</ul>

**Alfa'dan Kritik Değere**

**Kritik değer,** seçilen α değerine karşılık gelen test istatistiği değeridir.</br>

**α = 0,05 için sağ kuyruklu test:**</br>

<ul>
    <li>Kritik z değeri = 1.645</li>
    <li>Bu, sağında alanın tam %5'i olan z puanıdır</li>
</ul>

**α = 0,05 için iki kuyruklu test:**</br>

<ul>
    <li>Kritik z değerleri = ±1,96</li>
    <li>Bunlar her kuyrukta %2,5'e sahip olan z puanlarıdır (toplam %5)</li>
</ul>

**Kritik değerleri kullanan karar kuralı:**</br>

<ul>
    <li>Hesaplanan z > kritik z ise: H₀'yi reddet</li>
    <li>Hesaplanan z < kritik z ise: H₀'ı reddedemezsin</li>
</ul>

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# x ekseni
x = np.linspace(-4, 4, 400)

# Normal dağılım
y = norm.pdf(x)

# z-değeri
z_value = 2

plt.figure(figsize=(8, 4))

# Normal dağılım eğrisi
plt.plot(x, y, linestyle="--", linewidth=2, label="Normal Distribution")

# p-value alanını renklendir (z'nin sağ tarafı)
x_fill = np.linspace(z_value, 4, 200)
plt.fill_between(x_fill, norm.pdf(x_fill), color="orange", alpha=0.4, label="p-value area")

# z = 2 çizgisi
plt.axvline(z_value, color="black", linestyle="-", linewidth=1.5)
plt.text(z_value + 0.05, 0.05, "z = 2", fontsize=10)

# Eksen ve etiketler
plt.axhline(0, color="black", linewidth=0.8)
plt.xlim(-3.5, 3.5)
plt.xticks([-3, -2, -1, 0, 1, 2, 3])

plt.title("Normal Distribution & p-value for z = 2")

plt.legend()

# Alt açıklama
plt.text(2, -0.08, "<---> p-value (shaded area)", fontsize=9)

plt.tight_layout()
plt.show()
````

**Aynı Kararı Vermenin İki Yolu**

**Yöntem 1: p değerini α ile karşılaştırın**<br>

<ul>
    <li>Test istatistiğini hesapla → p değerini al → α ile karşılaştır</li>
    <li>Modern yaklaşım, yorumlaması daha kolay</li>
</ul>


**Yöntem 2: Test istatistiğini kritik değerle karşılaştırın**<br>

<ul>
    <li>Test istatistiğini hesaplayın → Kritik değerle karşılaştırın</li>
    <li>Geleneksel yaklaşım, hala yaygın</li>
</ul>

**İkisi de aynı! İkisi de aynı sonuca varıyor.**<br>

**Örnek:**</br>

Veriler:
●​ z = 2,0 </br>
●​ α = 0,05 </br>
●​ Sağ kuyruk testi </br>

Yöntem 1 (p-değeri yaklaşımı): </br>
●​ p-değeri = 0,0228 </br>
●​ 0,0228 < 0,05 </br>
●​ H₀'yi reddet </br>

Yöntem 2 (kritik değer yaklaşımı): </br>
●​ Kritik değer = 1,645 </br>
●​ 2,0 > 1,645 </br>
●​ H₀'yi reddet </br>



**T-İstatistik ve T-Dağılımını Anlamak**

t istatistiği z ile aynı şekilde çalışır, ancak σ'yu bilmediğinizde ek belirsizlik ortaya çıkar.</br>

**T-Dağılım Ailesi**

Z-dağılımının aksine (ki her zaman aynıdır), her serbestlik derecesi için farklı bir t-dağılımı vardır. </br>



````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, t

# x ekseni
x = np.linspace(-4, 4, 400)

# Dağılımlar
y_norm = norm.pdf(x, loc=0, scale=1)   # z ~ N(0,1)
df = 5
y_t = t.pdf(x, df=df)                  # t(df=5)

plt.figure(figsize=(8, 4))

# Eğriler
plt.plot(x, y_norm, label="z (normal)", linewidth=2)
plt.plot(x, y_t, linestyle="--", label=f"t (df={df})", linewidth=2)

# Eksen ve arka plan ayarları
plt.axhline(0, color="black", linewidth=0.8)
plt.axvline(0, color="black", linewidth=0.5, linestyle=":")

plt.xlim(-3.5, 3.5)
plt.xticks([-3, -2, 0, 2, 3])

plt.xlabel("x")
plt.ylabel("Yoğunluk")
plt.title("Normal dağılım vs t-dağılımı (df=5)")

plt.legend()

plt.show()

````

T dağılımı normal dağılıma göre daha kalın kuyruklara sahiptir. df arttıkça, t dağılımı → normal dağılım.

**T-Puanı Yorumlanması**
Z-puanı gibi t-puanı da "beklenenden kaç standart hata uzakta olduğunuzu" söyler.</br>

**Örnek:** t = 2.5 ve df = 10 </br>

**Bunun anlamı:** "Örneklem ortalaması, varsayılan ortalamanın 2,5 standart hata üzerindedir ve bu, standart sapma tahminimizdeki belirsizliği hesaba katar."</br>

**p-değerini elde etmek:** t-tablosunda df = 10'a bakılır veya bir yazılım kullanılır.</br>

df = 10 için sağ kuyruklu test:</br>

●​ α = 0,05'teki kritik t değeri yaklaşık 1,812'dir</br>
●​ 2,5 > 1,812 olduğundan → H₀'yi reddet</br>



**Serbestlik Dereceleri (df):**

Bu, yeni başlayanlar için en gizemli kavramlardan biridir.</br>

**Serbestlik derecesi =** Bir istatistiğin hesaplanmasında kullanılan bağımsız bilgi parçalarının sayısı.</br>

Ya da daha basiti: **Kaç değer "serbestçe değişebilir"?**</br>

**Örnekler**</br>

**Örnek 1: Örnek Ortalaması (Tek Örneklem T-Testi)**</br>

5 sayınız var. Size şunu söyleyeyim:</br>
●​ Ortalama 10 </br>
●​ Sayılardan dördü: 8, 9, 11, 12 </br>

**Soru:** Beşinci sayı nedir?</br>

**Çözüm:** Ortalama 10 ise, toplam 5 × 10 = 50 olmalıdır. Mevcut toplam = 8 + 9 + 11 + 12 = 40 Beşinci sayı = 50 - 40 = 10 olmalıdır.</br>

**Önemli nokta:** Ortalama ve 4 sayı bilindiğin, 5. sayının değişmesi mümkün değildir - belirlenmiştir!</br>

**Serbestlik Derecesi =** n - 1 = 5 - 1 = 4 </br>

Sadece 4 değer değişebilir. 5.si kısıtlanmıştır.</br>

**Örnek 2: İki Örneklemli T Testi**</br>

Grup 1: n₁ = 20 kişi Grup 2: n₂ = 25 kişi</br>

İki örneklem ortalaması hesaplarsınız (her grup için bir tane).</br>

df = (n₁ - 1) + (n₂ - 1) = 19 + 24 = 43 </br>
Veya eşdeğer olarak:</br>
df = n₁ + n₂ - 2 = 20 + 25 - 2 = 43</br>

Neden 2 çıkarılır? Çünkü 2 parametre (iki ortalama) tahmin edildi ve her biri  1 serbestlik derecesine mal oldu.</br>

**Örnek 3: Eşleştirilmiş T-Testi**</br>

15 çift ölçümü var (15 kişide öncesi/sonrası). 15 fark hesaplanıyor. Ardından bu farkların ortalamasını hesaplanıyor.</br>

df = n - 1 = 15 - 1 = 14 </br>

Tıpkı tek örneklemli bir test gibi! Çünkü farklılıklar "örneklem" olarak ele alınıyor.</br>

**Örnek 4: Ki-Kare Testi**</br>

**Uyum iyiliği testi: 6 kategoriniz var (zar yüzleri gibi).**</br>

**df = (number of categories) - 1 = 6 - 1 = 5**</br>

Neden? Toplam sayı ve 5 kategori sayısı biliniyorsa, 6. belirlenir. </br>

**Bağımsızlık testi (2×3 tablo): 2 satır, 3 sütun**</br>

**df = (rows - 1) × (columns - 1) = (2-1) × (3-1) = 1 × 2 = 2**</br>

Neden? Satır toplamları, sütun toplamları ve (satır-1)×(sütun-1) hücre verildiğinde diğer tüm hücreler belirlenir.</br>

**Örnek 5: ANOVA**</br>

Toplam N = 40 gözlem içeren k = 4 grup karşılaştırıldığında.</br>

Gruplar arası df: df₁ = k - 1 = 4 - 1 = 3 (4 grup ortalamasını tahmin ediyorsunuz, ancak bunlar genel ortalama ile sınırlandırılmıştır)</br>

Grup içi df: df₂ = N - k = 40 - 4 = 36 (Her grup ortalamasından 1 df kaybeder)</br>

**Serbestlik Derecesinin Önemi**</br>

<ol>
    <li>df değeri, istatistiksel testte kullandığın dağılımın şeklini belirler.
        <ol>
            <li>t-dağılımı df’ye bağlı olarak şekil değiştirir.</li>
            <li>df küçükse (ör. 5), dağılım daha yayvan ve kuyrukları daha şişmandır.</li>
            <li>df büyüdükçe (örn. 30), dağılım normal dağılıma yaklaşır.</br>
                ⇒ df → ∞ → t-dağılımı = normal dağılım gibi olur.
            </li>
        </ol>
    </li>
     <li>df, kritik değerleri değiştirir → bu da H₀’ı reddetmeyi etkiler.
        <ol>Örneğin α = 0.05, iki yönlü test için:
            <li>z-kritik (df = ∞) = 1,96</li>
            <li>t-kritik (df = 10) = 2,228</li>
            <li>t-kritik (df = 5) = 2,571</li>
            <li>t-kritik (df = 2) = 4,303</li>
        </ol>
         <p>👉 df azaldıkça kritik değer büyür</p>
         <p>👉 Kritik değer büyüdükçe H₀’ı reddetmek daha zorlaşır</p>
         <b>NOT</b>:Daha küçük bir df → daha fazla belirsizlik → daha geniş/şişman kuyruk → daha ekstrem bir test istatistiği gerektirir.
    </li>
     <li>df örneklem büyüklüğünü ve model karmaşıklığını yansıtır
        <ol>
            <li>Daha büyük sample size → daha büyük df → daha hassas (precise) tahmin</li>
            <li>Modelde daha fazla parametre tahmin edildikçe → df azalır → tahminler daha az güvenilir</li>
            Yani df hem kaç gözlemin var hem de modelde kaç parametre kullandığın konusunda bilgi taşır.
        </ol>
    </li>
    
</ol>

**Felsefi Anlam**</br>
Serbestlik dereceleri bilgiyi temsil eder.</br>

Verilerinizden bir parametreyi (örneğin bir ortalama) her tahmin ettiğinizde, bilgilerinizin bir kısmını "tüketirsiniz". Geriye kalanlar ise hipotezleri test etmek için kullanılabilir.</br>

Bunu para gibi düşünün:</br>
●​ n dolarla (n veri noktası) başlıyorsunuz.</br>
●​ Tahmin ettiğiniz her parametrenin maliyeti 1 dolardır.</br>
●​ Geriye kalan, çıkarımlar yapmak için gereken "satın alma gücünüz" olan df'nizdir.</br>


**Güven Aralıkları: Madalyonun Diğer Yüzü**</br>

Şimdi güven aralıklarından ve bunların hipotez testleriyle ilişkisinden bahsedelim.

**Güven Aralığı Nedir?**</br>

Bir güven aralığı, bir popülasyon parametresi için makul değerler aralığı verir.</br>

**Genel biçimi:**</br>

CI = (point estimate) ± (critical value) × (standard error)</br>

Ortalama İçin:</br>

$
CI = \bar{x} \pm t({\alpha/2,\, df}) \times \left(\frac{s}{\sqrt{n}}\right)
$</br>

**"%95 Güven"i Anlamak**</br>

**Yaygın Yanlış Yorumlama:** ❌ "Gerçek ortalamanın bu aralıkta olma olasılığı %95'tir."</br>

**Doğru Yorumlama:** ✓ "Bu işlemi birçok kez tekrarlarsak, oluşturduğumuz güven aralıklarının yaklaşık %95'i gerçek popülasyon ortalamasını içerecektir."</br>

**Frekansçı Yorum**</br>

Diyelim ki:</br>

1. Örnek alın
2. %95 GA hesaplayın
3. Bunu 100 kez tekrarlayın

**Sonuç:** Aralıklarınızın yaklaşık %95'i gerçek μ'yi içerecek, yaklaşık %5'i ise içermeyecektir.</br>

**Visual:**</br>

Gerçek μ = 50</br>

Örnek 1: [48 ════════ 52] ✓ μ İçerir</br>
Örnek 2: [47 ════════ 51] ✓ μ İçerir</br>
Örnek 3: [49 ════════ 53] ✓ μ İçerir</br>
Örnek 4: [51 ════════ 55] ✗ μ'yi kaçırır (zamanın yaklaşık %5'inde gerçekleşir)</br>
Örnek 5: [46 ════════ 50] ✓ μ İçerir</br>


**Örnek: Güven Aralığının Hesaplanması**</br>

**Senaryo:** 25 kişinin boyunu ölçüyorsunuz.</br>

●​ Sample mean: $\bar{x}$ = 170 cm</br>
●​ Sample SD: s = 10 cm </br>
●​ Sample size: n = 25 </br>

**İstenen:** Gerçek ortalama yükseklik μ için %95 güven aralığı.</br>

**Adım 1: Kritik t değerini bulun**</br>

●​ df = n - 1 = 24 </br>
●​ %95 CI için α = 0,05, dolayısıyla α/2 = 0,025 (iki kuyruklu)</br>
●​ t-tablosundan: t(0,025, 24) = 2,064 </br>

**Adım 2: Standart hatayı hesaplayın**</br>

SE = s / √n = 10 / √25 = 10 / 5 = 2</br>

**Adım 3: Hata payını hesaplayın**</br>
ME = t × SE = 2.064 × 2 = 4.128

**Adım 4: Aralığı oluşturun**</br>

CI = $\bar{x}$ ± ME </br>
CI = 170 ± 4.128 </br>
CI = (165.87, 174.13) </br>

**Yorum:** "Nüfusun gerçek ortalama boyunun 165,87 cm ile 174,13 cm arasında olduğundan %95 eminiz."</br>

**Derin Bağlantı: Güven Aralıkları ve Hipotez Testleri**</br>

İşte asıl ilginç nokta burada başlıyor: Güven aralıkları ve hipotez testleri aynı şeyi söylemenin iki farklı yoludur!</br>

**Eşdeğerlik**</br>

α önem düzeyinde iki taraflı bir test için:</br>

**Hipotez Testi Yaklaşımı:**</br>

<ul>
    <li>H₀: μ = μ₀</li>
    <li>Test istatistiğini hesapla</li>
    <li>p-değerini al</li>
    <li>Eğer p < α ise, H₀'yi reddedin</li>
</ul>

**Güven Aralığı Yaklaşımı:**</br>

<ul>
    <li>(1-α)% güven aralığını hesaplayın</li>
    <li>μ₀ aralıkta DEĞİLSE, H₀'yi reddedin</li>
    <li>μ₀ aralıktaysa, H₀'yi reddetmeyi başaramaz.</li>
</ul>

Aynı cevabı veriyorlar!</br>

**Örnek: Eşdeğerliği Görmek**</br>

**Senaryo:** μ = 165 cm olup olmadığını test edin. Daha önce, %95 CI'mız = (165,87, 174,13) idi.</br>

**Hipotez Testi:*/br>*

<ul>
    <li>H₀: μ = 165</li>
    <li>Hₐ: μ ≠ 165</li>
    <li>α = 0.05</li>
</ul>

t'yi hesaplayın:</br>

t = ($\bar{x}$ - μ₀) / (s/√n)</br>

t = (170 - 165) / 2 </br>
t = 2.5 </br>

α = 0,05, df = 24'teki kritik t değeri: ±2,064 2,5 > 2,064 olduğundan → H₀'yi reddet </br>

Güven Aralığı Yöntemi: %95 GA'mız (165.87, 174.13)'tür. Bu aralıkta μ₀ = 165 mi? HAYIR! (165 < 165.87) → H₀'yi Reddet </br>

**Hepsini Bir Araya Getirelim:**</br>

**Tam Hipotez Test Süreci**</br>

**Senaryo:** Bir kahve şirketi, poşetlerinin 500 gr kahve içerdiğini iddia ediyor. Siz de poşetlerin yetersiz olduğundan şüpheleniyorsunuz. 16 poşet seçtiniz.</br>

**Veriler:**</br>

<ul>
    <li>n = 16 poşet</li>
    <li>$\bar{x}$ = 495g (sample mean)</li>
    <li>s = 12g (örnek SD)</li>
</ul>


**Adım 1: Hipotezleri belirtin ve α'yı seçin**</br>

<ul>
    <li>H₀: μ ≥ 500 g (eksik doldurma değil)</li>
    <li>Hₐ: μ < 500 g (eksik doldurma)</li>
    <li>α = 0,05 (önceden seçilmiştir)</li>
    <li>Bu sol kuyruklu bir testtir</li>
</ul>


**Adım 2: Test istatistiğini hesaplayın**</br>

t = ($\bar{x}$ - μ₀) / (s/√n)</br>
t = (495 - 500) / (12/√16)</br>
t = -5 / (12/4)</br>
t = -5 / 3</br>
t = -1.667</br>

**Bu t-puanı ne anlama geliyor:** "Örneklem ortalaması, iddia edilen ortalamanın 1,667 standart hata altındadır."</br>

**Adım 3: Serbestlik derecelerini belirleyin**</br>

df = n - 1 = 16 - 1 = 15</br>

**Adım  4: Kritik değeri bulun (geleneksel yaklaşım)**</br>

Sol kuyruklu test için α = 0,05, df = 15: Kritik t değeri = -1,753 </br>

**Kritik değer kullanılarak karar:** t = -1,667, -1,753'ten daha mı uç değerde? Hayır! -1,667 > -1,753 (sıfıra daha yakın) → **H₀ reddedilemedi**</br>

**Adım 5: p-değerini bulun (modern yaklaşım)**</br>

t = -1,667, df = 15 için t-tablosu veya yazılım kullanıldığında: p-değeri ≈ 0,058 </br>

**p-değeri kullanılarak karar:** p = 0,058 < α = 0,05 mi? Hayır! 0,058 > 0,05 → **H₀ reddedilemedi** </br>

**Her iki şekilde de aynı sonuç!**</br>


**Adım 6: Güven aralığını hesaplayın**</br>

df = 15 olan %95 CI için: t-kritik (iki kuyruklu) = 2,131 </br>

$
CI = \bar{x} \pm t \times \left(\frac{s}{\sqrt{n}}\right)
$</br>

CI = 495 ± 2.131 × 3</br>
CI = 495 ± 6.393</br>
CI = (488.61, 501.39)</br>

**CI'nin Yorumlanması:** "Gerçek ortalama ağırlığın 488,61 g ile 501,39 g arasında olduğundan %95 eminiz." </br>

**Dikkat:** Bu aralık 500g'ı kapsamaktadır! Bu, H₀'yi reddetmeme kararımızı doğrulamaktadır.</br>

**Adım 7: Sonucu tamamlayın**</br>

"16 poşetlik örneğimize dayanarak (ortalama = 495 g, SD = 12 g), poşetlerin 500 g'ın altında yetersiz doldurulduğu sonucuna varmak için α = 0,05 düzeyinde yeterli kanıtımız yok (t = -1,667, df = 15, p = 0,058). %95 güven aralığı (488,61 g, 501,39 g), iddia edilen 500 g değerini de içermektedir. Örneklem ortalaması 500 g'ın altında olsa da, gerçek ortalama 500 g ise bu farkın tesadüfen ortaya çıkması makul bir olasılıktır."
