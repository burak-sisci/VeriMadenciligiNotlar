## Hipotez Testine Giriş ve Hipotez Testinin Temelleri

**Hipotez Testi Neden Gereklidir**

Diyelim ki bir kahve zincirinin müdürüsünüz. Kahve çekirdeği tedarikçiniz bir gün sizi arayıp, "Hey! Kavurma sürecimizi geliştirdik. Artık kahve çekirdeklerimiz size daha zengin bir tat verecek ve müşterileriniz bayılacak!" diyor.</br>

Kulağa hoş geliyor, değil mi? Ama sorun şu: Tedarikçinin doğruyu söyleyip söylemediğini nasıl anlarsınız? </br>

Sadece onların sözüne güvenemezsin çünkü:

<ul>
    <li>Belki de size daha pahalı çekirdekler satmaya çalışıyorlardır.</li>
    <li>Belki de "iyileştirme" gerçek kahve dükkanlarında değil sadece kendi laboratuvarlarında işe yarıyordur.</li>
    <li>Belki de birkaç test partisinin daha iyi sonuç vermesi sadece şans eseridir.</li>
</ul>

Peki ne yaparsınız? Kendiniz denemeye karar verirsiniz.</br>

"Yeni geliştirilmiş" çekirdeklerden alıp 50 müşteri için kahve demliyorsunuz. Müşterilerden puanlamalarını istiyorsunuz. Ardından bu puanları, eski çekirdeklerinizle normalde aldığınız puanlarla karşılaştırıyorsunuz.</br>

Ama burada işler zorlaşıyor... </br>

Diyelim ki eski çekirdekleriniz genelde 10 üzerinden 7.5 puan alıyor ve yeni çekirdeklerinize de 50 müşteriniz ortalama 7.8 puan veriyor. </br>

**Soru:** Bu, yeni çekirdeklerin gerçekten daha iyi olduğu anlamına mı geliyor? Yoksa o gün keyfi yerinde olan 50 müşteriye mi hizmet verdiniz? </br>

Hipotez testinin çözdüğü sorun tam da budur. "Bu fark gerçek mi, yoksa sadece rastgele bir şans mı?" sorusuna cevap bulmanıza yardımcı olur. </br>




**Hipotez Testinin Çözdüğü Temel Sorun**

Her gün, her alanda benzer durumlarla karşılaşıyoruz:<br>

**Tıpta:** Bir ilaç şirketi, yeni diyabet ilacının kan şekerini mevcut ilaçlardan daha iyi düşürdüğünü iddia ediyor. Peki, ilacın işe yaradığını ve hastalardaki rastgele bir değişim olmadığını nasıl anlıyorsunuz? </br>

**Pazarlamada:** Web sitenizin buton rengini maviden kırmızıya değiştiriyorsunuz ve satışlar %3 artıyor. Harika! Ama durun bakalım... bunun sebebi buton rengi mi, yoksa zaten satışlar açısından iyi bir hafta mıydı?</br>

**Üretimde:** Bir fabrika, ortalama 1000 saat dayanması gereken ampuller üretiyor. 30 ampulü test ediyorsunuz ve ortalama 980 saat dayanıyorlar. Üretim hattını kapatıp araştırma yapmalı mısınız, yoksa bu normal bir değişim mi? </br>

**Ortak Nokta:** Tüm bu durumlarda, bazı veriler (değerlendirmeler, satışlar, ampul ömürleri) görürüz ve şunu bilmek isteriz: "Gördüğüm şey gerçek bir desen mi, yoksa sadece rastgele bir gürültü mü?" </br>

İşte bu noktada hipotez testi devreye giriyor. Bu, ne kadar yanılma riski aldığımızı tam olarak bilerek bu kararları almanın resmi ve matematiksel bir yoludur. </br>





**İstatistiksel Çıkarımı Anlamak: Gördüklerimizle Bilmek İstediklerimiz Arasındaki Köprü**

Şimdi, hipotez testine özel olarak dalmadan önce, bunun bir parçası olan daha büyük bir kavramı anlamamız gerekiyor: istatistiksel çıkarım. </br>

Bunu başka bir hikayeyle anlatayım. </br>

**Çorba Benzetmesi**

Bir parti için kocaman bir tencere çorba pişiriyorsunuz - belki 10 litre. Yeterince tuz içerip içermediğini merak ediyorsunuz, ama bunu öğrenmek için 10 litrenin tamamını içemezsiniz! </br>

Ne yaparsınız? Bir kaşık alır, tadına bakar ve tencerenin tamamına daha fazla tuz gerekip gerekmediğine karar verirsiniz.</br>

O kaşık dolusu numuneniz. Tüm kap sizin popülasyonunuz. Peki, kaşığı tatma ve kap dolusu hakkında karar verme süreci? İşte bu istatistiksel çıkarımdır. </br>

Ama mesele şu: Ya tuz eşit şekilde karışmasaydı? Ya o bir kaşık dolusu, tencerenin geri kalanından daha fazla tuz içerseydi? Daha az tuz eklerdiniz ve tencerenin tamamı tuzsuz kalırdı! </br>

Temel zorluk şudur: Sadece bir kaşık (örnek) ile tüm pot (popülasyon) hakkında güvenilir kararlar nasıl alabiliriz?</br> 

**İstatistiksel Çıkarım Nedir?**

İstatistiksel çıkarım, daha küçük bir gruptan (örneklem) alınan bilgilere dayanarak büyük bir grup (popülasyon) hakkında sonuçlar çıkarma bilimidir.</br>

Bunu yapmanın iki temel yolu vardır:</br>

<ol>
    <li>
        **Tahmin:** "Müşterilerimizin ortalama yaşı kaçtır?"
        <ul>
            <li>1.000 müşteriyi ankete tabi tutuyoruz ve ortalama yaşlarını hesaplıyoruz</li>
            <li>Sonra diyoruz ki, "TÜM müşterilerimizin ortalama yaşı muhtemelen bu rakam civarındadır, biraz eksik veya fazla."</li>
        </ul>
    </li>
    <li>
        **Hipotez Testi:** "Yeni eğitim programımız çalışan verimliliğini artırdı mı?"
         <ul>
            <li>Eğitim öncesi ve sonrası verimliliği ölçüyoruz</li>
            <li>O zaman soruyoruz: "Bu gelişme gerçek mi, yoksa sadece rastgele bir şans mı?"</li>
        </ul>
    </li>
</ol>

Şöyle düşünün:</br>
<ul>
    <li> <b>Tahmin etmek</b>, "Cevap nedir?" diye sormak gibidir.</li>
    <li> <b>Hipotez testi</b>, "Bu cevap doğru mu?" diye sormak gibidir.</li>
</ul>



**Popülasyon ve Örneklem: Bu Ayrımın Önemi**

**Popülasyon Nedir?**

Nüfus, önemsediğiniz grubun tamamıdır; istisnasız her bir üye.</br>

Örnekler:</br>
<ul>
    <li>Şirketinizin 5 milyon müşterisinin tamamı</li>
    <li>Bu yıl üniversitenize kayıtlı her öğrenci</li>
    <li>Belirli bir göldeki tüm balıklar</li>
    <li>Fabrikanız tarafından bu ay üretilen her civata</li>
    <li>Hindistan'da akıllı telefon kullanan tüm yetişkinler</li>
</ul>

Asıl bilmek istediğimiz şey popülasyondur. Popülasyondaki herkesi veya her şeyi ölçebilseydik, istatistiklere hiç ihtiyacımız olmazdı! Sadece istediğimiz tam değerleri hesaplardık. </br>

Ancak sorun şu: Nüfusun tamamını neredeyse hiçbir zaman ölçemeyiz.</br>

Niçin? </br>

<ul>
    <li><b>Maliyet:</b> 5 milyon müşteriyi test etmek çok pahalıya mal olur</li>
    <li><b>Zaman:</b> Herkesi ölçmeyi bitirdiğinizde durum değişmiş olur</li>
    <li><b>İmkansızlık:</b> Nüfus sonsuz olabilir (gelecekteki tüm müşteriler gibi)</li>
    <li><b>Yıkım:</b> Bazı testler ölçtüğünüz şeyi yok eder (arabaların çarpışma testleri veya pillerin ne kadar süre dayanacağının test edilmesi gibi)</li>
</ul>

**Örneklem Nedir?**

Örneklem, popülasyondan seçilen daha küçük bir gruptur. </br>
Bu, bir çuvaldan bir avuç pirinç alıp pişip pişmediğini kontrol etmek gibi. Bütün çuvalı yemenize gerek yok! </br>

Örnekler:</br>
<ul>
    <li>5 milyon müşteri arasından rastgele seçilen 1.000 müşteri</li>
    <li>Üniversitenizden 300 öğrenci ankete katıldı</li>
    <li>Gölden yakalanan ve ölçülen 50 balık</li>
    <li>Bugün üretilen binlerce civatadan arasından test edilen 100 civata</li>
    <li>Milyonlarca akıllı telefon kullanıcısı arasından ankete katılan 2.000 akıllı telefon kullanıcısı</li>
</ul>

**Ana Görüş**

İstatistikleri güçlü kılan şey şudur: Örneğiniz doğru (rastgele) seçilirse, popülasyonu makul ölçüde iyi yansıtacaktır.</br>

Bir düşünün: Bir deste kartı doğru şekilde karıştırıp beş kart çekerseniz, bu beş kart destede ne olduğunu anlamanıza yardımcı olur. Her zaman tam olarak doğru olmayabilir, ama çok da yanlış olmaz.



**Popülasyonlar ve Örneklemler Hakkında Nasıl Konuşulur: Notasyon**

İstatistikçiler, popülasyon değerleri ile örneklem değerleri için farklı semboller kullanırlar. Bu biraz titizlik gibi görünebilir, ancak aslında çok önemlidir çünkü hangisiyle uğraştığımızı hatırlatır.</br>

**For Means (Averages):**

<ol>
    <li>
        **Popülasyon Ortalaması:μ
        <ul>
            <li>Bu, bilmek istediğimiz ancak genellikle ölçemediğimiz GERÇEK ortalamadır</li>
            <li>Örnek: μ = Hindistan'daki TÜM yetişkinlerin ortalama boyu</li>
        </ul>
    </li>
    <li>
        **Örneklem Ortalaması:** $\bar{x}$ (x-bar)
         <ul>
            <li>Bu, örneğimizden gerçekten hesapladığımız ortalamadır</li>
            <li>Örneğin $\bar{x}$ = Ölçtüğümüz 1.000 yetişkinin ortalama boyu</li>
        </ul>
    </li>
</ol>


**Standart Sapmalar İçin:**

<ol>
    <li>
        **Popülasyon Standart Sapması:σ
        <ul>
            <li>Nüfusun tamamı ne kadar yayılmış?</li>
        </ul>
    </li>
    <li>
        **Örneklem Standart Sapması:** s
         <ul>
            <li>Örneklemimiz ne kadar yayılmış?</li>
        </ul>
    </li>
</ol>

**Oranlar (Yüzdeler) İçin:**

<ol>
    <li>
        **Popülasyon Oranı:p
        <ul>
            <li>Örnek: p = memnun olan tüm müşterilerin GERÇEK yüzdesi</li>
        </ul>
    </li>
    <li>
        **Örneklem Oranı:** $\hat{p}$ (p-hat)
         <ul>
            <li>Örnek: $\hat{p}$ = örneklemimizde memnun olan müşterilerin yüzdesi</li>
        </ul>
    </li>
</ol>

**Boyut(Size) İçin:**
<ul>
    <li><b>Popülasyon boyutu:</b> N </li>
    <li><b>Popülasyon ortalama geliri:</b>  </li>
</ul>


**Senaryo:** Bangalore'daki tüm yazılım mühendislerinin ortalama gelirini bilmek istiyorsunuz.

<ul>
    <li><b>Popülasyon:</b>  Bangalore'daki tüm yazılım mühendisleri (N = 500.000 diyelim) </li>
    <li><b>Örneklem ortalama geliri:</b> μ = ??? (Bu bilinmiyor - öğrenmek istediğimiz şey bu!) </li>
    <li><b>Örneklem:</b> Rastgele seçilen 200 yazılım mühendisi (n = 200) </li>
    <li><b>Örneklem ortalama geliri:</b>  $\bar{x}$ = ₹12.5 lakhs, μ'nin muhtemelen ₹12,5 lakh civarında olduğunu tahmin ediyoruz (tam olarak değil - biraz belirsizlik var). </li>
</ul>




**Sıfır ve Alternatif Hipotezler: Hipotez Testinin Özü**

**Bir Benzetmeyle Başlayalım: Mahkeme Duruşması**

Diyelim ki biri suçlanıyor ve yargılanıyor. Adalet sistemi nasıl işliyor? <b>

**Varsayılan Durum:** Kişinin suçluluğu kanıtlanana kadar masum olduğu varsayılır.</br>
**Savcının Görevi:** Kişinin makul şüphenin ötesinde suçlu olduğunu kanıtlayacak delilleri sunmak.</br>

**Sonuç:**

<ul>
    <li>Eğer deliller yeterince güçlüyse → Kişi suçlu ilan edilir</li>
    <li>Eğer deliller YETERİNCE güçlü DEĞİLSE → Kişi "suçsuz" ilan edilir (dikkat: "masum" demiyoruz!)</li>
</ul>

<i>Neden "Masum" yerine "Suçsuz"? Çünkü suçluluğu kanıtlayamamak masumiyeti kanıtlamaz. Bu sadece, mahkûmiyet için yeterli delil olmadığı anlamına gelir. Kişi belki masumdur, belki de değildir - elimizde yeterli delil yok.</i></br>

**Hipotez testi tam olarak böyle çalışır!**

**Sıfır Hipotezi (H₀): Mevcut Durum**

Sıfır hipotezi, "masumiyet varsayımı" pozisyonuna benzer. Başlangıçta varsayılan varsayımımız budur.</br>

<ul>
    <li>Sembol: H₀ ("H-yokluk" veya "H-sıfır" olarak okunur)</li>
    <li>Anlamı: "Hiçbir özel şey olmuyor" veya "Hiçbir etki" veya "Hiçbir fark yok"</li>
    <li>
        Örnekler:
        <ul>
            <li>"Yeni ilaç eskisinden daha iyi etki göstermiyor"</li>
            <li>"Madeni para adildir (yazı gelme olasılığı %50-%50)"</li>
            <li>"Müşteri memnuniyeti ortalamamız geçen yılla aynı"</li>
        </ul>
    </li>

</ul>

**Önemli Nokta:** Sıfır hipotezi HER ZAMAN bir eşitlik işareti (=, ≤ veya ≥) içerir. </br>

Neden? Çünkü bir şeyin belirli bir değere eşit olup olmadığını veya belirli bir aralıkta olup olmadığını test ediyoruz.


**Alternatif Hipotez (Hₐ veya H₁): Kanıtlamaya Çalıştığımız Şey**

Alternatif hipotez, "suçlu" iddiasına benziyor. Kanıt bulmaya çalıştığımız şey bu. </br>

<ul>
    <li>Sembol: Hₐ veya H₁ ("H-a" veya "H-bir" olarak okunur)</li>
    <li>Anlamı: "Özel bir şey oluyor" veya "Bir etki var" veya "Bir fark var"</li>
    <li>
        Örnekler:
        <ul>
            <li>"Yeni ilaç eskisinden daha iyi etki ediyor"</li>
            <li>"Madeni para taraflıdır (50-50 değil)"</li>
            <li>"Müşteri memnuniyetimiz geçen yıla göre değişti"</li>
        </ul>
    </li>

</ul>


**Önemli Nokta:** Alternatif hipotez bir eşitsizlik işareti (≠, <, veya >) içerir.


**Bir Araya Getirme: Gerçek Örnekler**

**Örnek 1:** Kahve Çekirdekleri (İlk örneği hatırlayın?)</br>

**Durum:** Tedarikçiniz yeni çekirdeklerin daha iyi olduğunu iddia ediyor. Eski çekirdekleriniz ortalama 7.5 puan aldı. Yeni çekirdeklerin gerçekten daha iyi olup olmadığını test etmek istiyorsunuz.</br>

Hipotezler:

<ul>
    <li>H₀: μ = 7.5 (Yeni çekirdeklerin ortalama puanı eski çekirdeklerle aynıdır)</li>
    <li>Hₐ: μ > 7.5 (Yeni çekirdeklerin ortalama puanı daha yüksektir)</li>
</ul>

Düz bir şekilde ifade edilirse:

<ul>
    <li>H₀ diyor ki: "Hiçbir şey değişmedi, çekirdekler aynı"</li>
    <li>Hₐ diyor ki: "Yeni çekirdekler aslında daha iyi"</li>
</ul>


**Örnek 2:** Fabrika Kalite Kontrolü</br>

**Durum:** Fabrikanız ortalama çapı 5 mm olması gereken civatalar üretiyor. Bunların standartlara uygun olmayabileceği konusunda endişeleniyorsunuz..</br>

Hipotezler:

<ul>
    <li>H₀: μ = 5mm (Ortalama çap doğrudur)</li>
    <li>Hₐ: μ ≠ 5mm (Ortalama çap DOĞRU DEĞİLDİR - çok büyük veya çok küçük olabilir)</li>
</ul>

Düz bir şekilde ifade edilirse:

<ul>
    <li>H₀ diyor ki: "Her şey yolunda, üretim hedefte"</li>
    <li>Hₐ diyor ki: "Bir şeyler ters gidiyor, hedeften saptık"</li>
</ul>

**Örnek 3:** Tıbbi Çalışma</br>

**Durum:** Yeni bir ilacın kan basıncını düşürüp düşürmediğini test ediyorsunuz. Şu anda hasta grubunuzda ortalama kan basıncı 140 mmHg..</br>

Hipotezler:

<ul>
    <li>H₀: μ ≥ 140 mmHg (İlaç kan basıncını düşürmez)</li>
    <li>Hₐ: μ < 140 mmHg (İlaç kan basıncını düşürüyor)</li>
</ul>

Düz bir şekilde ifade edilirse:

<ul>
    <li>H₀ diyor ki: "İlaç işe yaramıyor"</li>
    <li>Hₐ diyor ki: "İlaç işe yarıyor ve kan basıncını düşürüyor"</li>
</ul>


**Neden Her Zaman H₀ Test Ediliyor da Hₐ test edilmiyor?**
Bu durum yeni başlayanların kafasını karıştırır, bu yüzden dikkatlice açıklanmalıdır.</br>.
Hₐ'yi asla doğrudan kanıtlamayız. Bunun yerine, H₀'yi çürütmeye çalışırız.</br>
Mahkeme benzetmesini tekrar düşünün: Savcı masumiyetin yanlış olduğunu kanıtlamaz; suçluluğun doğru olduğunu kanıtlar. Benzer şekilde:

<ul>
    <li>Veri topluyoruz</li>
    <li>Şunu kontrol ediyoruz: "H₀ doğru olsaydı, verilerim ne kadar sıra dışı olurdu?"</li>
    <li>Veriler H₀ altında ÇOK sıra dışıysa, "H₀ muhtemelen yanlıştır" diyor ve reddediyoruz</li>
    <li>Veriler H₀ altında o kadar sıra dışı değilse, "H₀'yi reddetmek için yeterli kanıt yok" diyoruz.</li>
</ul>


**Önemli:** ASLA "H₀'yi kabul ediyoruz" demiyoruz. Sadece "H₀'yi reddedemiyoruz" diyoruz.</br>

Bu tuhaf dil neden? Çünkü bir şeye karşı kanıt bulamamak, onun doğru olduğunu kanıtlamaz! </br>

**Benzetme:** Anahtarlarınızı 5 dakika boyunca arayıp mutfakta bulamazsanız, bu onların mutfakta olmadığını mı kanıtlar? Hayır! Belki yeterince dikkatli bakmadınız. Belki de açmadığınız bir çekmecededirler. "Bulamamak" ≠ "kesinlikle orada değiller."

Benzer şekilde, "H₀'ye karşı kanıt bulunamaması" ≠ "H₀ kesinlikle doğrudur."

### Tek Kuyruklu ve Çift Kuyruklu Testler: Hangi Yönü Önemsiyorsunuz?

Şimdi vermeniz gereken önemli bir karardan bahsedelim: Herhangi bir yöndeki değişimi mi, yoksa sadece belirli bir yöndeki değişimi mi önemsiyorsunuz? </br>

**Termometre Benzetmesi**

Diyelim ki bir hastanın vücut sıcaklığını ölçüyorsunuz, vücut sıcaklığı 37°C (normal).</br>

**Senaryo A:** Sıcaklığın değişip değişmediğini (ateş veya hipotermi) bilmek istiyorsunuz.

<ul>
    <li>37°C'den (ateş) DAHA YÜKSEK olup olmadığını önemsiyorsunuz</li>
    <li>37°C'den (hipotermi) DAHA DÜŞÜK olup olmadığını da önemsiyorsunuz</li>
    <li>Bu İKİ KUYRUKLU(TWO-TAILED) BİR TESTTİR</li>
</ul>

**Senaryo B:** Sadece ateşle mi ilgileniyorsunuz?

<ul>
    <li>Sadece 37°C'den DAHA YÜKSEK olup olmadığını önemsiyorsun</li>
    <li>Daha düşük olması umurunuzda değil (aslında daha düşük olması iyi bir haber!)</li>
    <li>Bu TEK KUYRUKLU (ONE-TAILED) bir TESTTİR (özellikle SAĞ KUYRUKLU (RIGHT-TAILED))</li>
</ul>

**Senaryo C:** Soğuk bir ortamda hipotermiden endişe duyuyorsunuz

<ul>
    <li>Sadece 37°C'den DÜŞÜK olup olmadığını önemsiyorsun</li>
    <li>Daha yüksek sıcaklık bir endişe kaynağı değil</li>
    <li>Bu TEK KUYRUKLU (ONE-TAILED) bir TESTTİR (özellikle SOL KUYRUKLU (LEFT-TAILED))</li>
</ul>


**Çift Taraflı Testler: Her İki Yöndeki Değişim Önemlidir**

**Ne zaman kullanılmalı:** Yönden bağımsız olarak HERHANGİ bir farklılığa önem verdiğinizde.</br>

Hipotezler nasıl yazılır:
<ul>
    <li>H₀: μ = [bir değer]</li>
    <li>Hₐ: μ ≠ [bir değer] ("≠" işaretine dikkat edin)</li>
</ul>



**Gerçek Örnek - Kalite Kontrol:**

500 gr'lık şeker poşetleri üretiyorsunuz. Tam olarak 500 gr olmasını istiyorsunuz - ne daha fazla ne daha az.

<ul>
    <li>H₀: μ = 500g (torbalar ortalama olarak tam 500g ağırlığındadır)</li>
    <li>Hₐ: μ ≠ 500g (torbalar ya az doldurulmuştur ya da aşırı doldurulmuştur)</li>
</ul>

Her iki yönün de önemi nedir:

<ul>
    <li>Eğer μ < 500g ise: Müşterileri aldatıyorsunuz demektir (yasal sorun!)</li>
    <li>μ > 500g ise: Bedava şeker vererek para kaybediyorsunuz</li>
</ul>


**Gerçek Örnek - Tıbbi Cihaz:**

Kan şekeri ölçüm cihazı doğru ölçümler vermelidir (yukarı veya aşağı yönlü bir sapma olmamalıdır).

<ul>
    <li>H₀: μ = [gerçek glikoz seviyesi]</li>
    <li>Hₐ: μ ≠ [gerçek glikoz seviyesi]</li>
</ul>

Her iki yönün de önemi nedir:

<ul>
    <li>Çok yüksek okuma: Hasta çok fazla insülin almış olabilir (tehlikeli!)</li>
    <li>Çok düşük okuma: Hasta yeterli insülin almıyor olabilir (bu da tehlikelidir!)</li>
</ul>

**Tek Yönlü Testler: Sadece Tek Yön Önemlidir**

**Sağ Kuyruk Testi (Üst Kuyruk)**

**Ne zaman kullanılır:** Sadece bir şeyin artıp artmadığı veya bir değerden büyük olup olmadığıyla ilgilendiğinizde.

Hipotezler nasıl yazılır:
<ul>
    <li>H₀: μ ≤ [bir değer]</li>
    <li>Hₐ: μ > [bir değer] (">" işaretine dikkat edin)</li>
</ul>

**Gerçek Örnek - İyileşmenin Kanıtlanması:**

Şirketinizin eğitim programı verimliliği artırdığını iddia ediyor. Eğitimden önce ortalama verimlilik saatte 50 birimdi.
<ul>
    <li>H₀: μ ≤ 50 (eğitim yardımcı olmuyor veya durumu daha da kötüleştiriyor)</li>
    <li>Hₐ: μ > 50 (eğitim aslında üretkenliği artırır)</li>
</ul>

Neden tek bir yön önemlidir: Sadece üretkenliğin ARTMALI olup olmadığıyla ilgilenirsiniz. Aynı kalsa da azalsa da, her iki durumda da eğitim başarısız olmuştur.


**Gerçek Örnek - Güvenlik Standardı::**

Bir yapı malzemesi EN AZ 5000 PSI basınca dayanıklı olmalıdır.
<ul>
    <li>H₀: μ ≤ 5000 PSI (malzeme standardı karşılamıyor)</li>
    <li>Hₐ: μ > 5000 PSI (malzeme standardı karşılıyor/aşıyor)</li>
</ul>

Neden sadece bir yön önemli: Tek yapmanız gereken 5000 PSI'dan DAHA GÜÇLÜ olduğunu kanıtlamak. Daha zayıf olmanın kabul edilemez olduğu zaten biliniyor.


**Sol Kuyruk Testi (Alt Kuyruk)**

**Ne zaman kullanılır:** Sadece bir şeyin azalması veya bir değerden az olmasıyla ilgilendiğinizde.

Hipotezler nasıl yazılır:
<ul>
    <li>H₀: μ ≥ [bir değer]</li>
    <li>Hₐ: μ < [bir değer] ("<" işaretine dikkat edin)</li>
</ul>

**Gerçek Örnek - Maliyetleri Azaltma:**

Şirketiniz üretim maliyetlerini düşürmek için yeni bir süreç uyguluyor. Mevcut maliyet birim başına 50 ABD dolarıdır.

<ul>
    <li>H₀: μ ≥ 50 $ (yeni süreç maliyetleri azaltmıyor)</li></br>
    <li>Hₐ: μ < 50 $ (yeni süreç aslında maliyetleri düşürüyor)</li>
</ul>

Neden tek bir yön önemlidir: Sadece maliyetlerin AZALMASINI önemsiyorsunuz. Maliyetler aynı kalırsa veya artarsa, yeni süreç amacına ulaşamamış demektir.


**Gerçek Örnek - Kirliliğin Azaltılması**

Yeni bir filtre kirliliği 100 ppm'nin altına düşürmelidir.
<ul>
    <li>H₀: μ ≥ 100 ppm (filtre yeterince iyi çalışmıyor)</li>
    <li>Hₐ: μ < 100 ppm (filtre kirliliği başarıyla azaltır)</li>
</ul>

Neden sadece bir yön önemlidir: Kirliliğin DAHA DÜŞÜK olduğunu göstermeniz yeterlidir. Daha yüksek kirlilik amacı boşa çıkarır.

**Bunu Düşünmenin Görsel Bir Yolu**

Varsayalım ki, hipotez ettiğiniz değerde bir sayı doğrusu üzerinde duruyorsunuz (diyelim ki μ = 100).

**İki taraflı:** Örneklem ortalamanızın her iki yönde de 100'den çok uzakta olması sizi endişelendiriyor.</br>
Çok düşük ← [H₀'yi Reddet] | 100 | [H₀'yi Reddet] → Çok yüksek

**Sağ kuyruklu:** Sadece örneklem ortalamanızın 100'ün çok üzerinde olup olmadığını önemsiyorsunuz.
Önemi yok ← | 100 | [H₀'yi reddet] → Çok yüksek!

**Sol kuyruklu:** Sadece örneklem ortalamanızın 100'ün çok altında olup olmadığını önemsiyorsunuz.
Çok düşük! ← [H₀'yi Reddet] | 100 | → Umurumda değil

**Hipotez Testinin Mantıksal Akışı**

Şimdi öğrendiğimiz her şeyi baştan sona eksiksiz bir örnekle bağlayayım.</br>

**Tam Örnek: Kahve Dükkanı Yeniden Ziyaret Edildi**

**Durum:** Kahve dükkanımızı hatırlıyor musunuz? Müşterilerden 10 üzerinden ortalama 7.5 puan alan kahve çekirdekleri kullanıyorsunuz (bu, yılların verilerine dayanıyor, bu yüzden 7.5 puana çok güveniyorsunuz). Tedarikçiniz artık daha iyi puan alacağını iddia ettiği "premium çekirdekler" sunuyor. Ancak premium çekirdeklerin fiyatı %20 daha pahalı! Değiştirmeli misiniz?

**Adım 1:Soruyu Çerçeveleyin** Premium çekirdekleri 40 müşteriyle test etmeye ve onların puanlarını almaya karar veriyorsunuz.
**Adım 2: Hipotezlerinizi Oluşturun**
Premium çekirdeklerin DAHA İYİ olduğunu kanıtlamaya çalışıyorsun, yani:
<ul>
    <li>H₀: μ ≤ 7.5 (Premium çekirdekler DAHA İYİ DEĞİLDİR; aynı veya daha kötü)</li>
    <li>Hₐ: μ > 7.5 (Premium çekirdekler DAHA İYİDİR)</li>
</ul>
Bu bir SAĞ KUYRUKLU testtir çünkü yalnızca derecelendirmelerin iyileşip iyileşmediğini önemsiyorsunuz.

**Adım 3: Örnek Verilerinizi Toplayın** Premium çekirdekleri n = 40 müşteriye servis ediyorsunuz ve onların puanlarını topluyorsunuz.

Diyelim ki hesaplıyorsunuz:

<ul>
    <li>Örneklem ortalaması. $\bar{x}$ = 7.9</li>
    <li>Örnek standart sapması: s = 0.8</li>
</ul>

**Adım 4: Temel Soru** "Eğer premium çekirdekler gerçekten normal çekirdeklerle aynıysa (H₀ doğru), sadece rastgele şans eseri 7.9'luk bir gruptan elde edilme olasılığım nedir?"

Eğer cevap "çok düşük ihtimal" ise, o zaman H₀'yi reddedersiniz ve premium çekirdeklerin gerçekten daha iyi olduğu sonucuna varırsınız.</br>
Eğer cevap "oldukça muhtemel" veya "bir nebze mümkün" ise, o zaman H₀'yi reddedemezsiniz - belki de 7.9 sadece şanstı!

**Adım 5: Karar Verin** (Bu hesaplamanın tam olarak nasıl yapıldığını sonraki bölümlerde öğreneceğiz, ancak şimdilik mantığını anlayın)

"Eğer H₀ doğru olsaydı, bu uç noktadaki veriyi elde etme şansı nedir?" diye soran bir olasılık (p-değeri) hesaplayacaksınız.

<ul>
    <li>p değeri küçükse (örneğin 0,02 = %2): "Vay canına, eğer H₀ doğru olsaydı, bu gerçekten nadir olurdu! Yani H₀ muhtemelen yanlıştır." → <b>H₀'yi reddet</b></li>
    <li>p değeri büyükse (örneğin 0,35 = %35): "H₀ doğru olsaydı, bu oldukça sık gerçekleşebilirdi. Verilerim o kadar da sıra dışı değil." → H₀'yi reddememek</li>
</ul>

**Adım 6: Bağlam İçinde Sonuç**

<ul>
    <li>H₀'yi reddettiyseniz: "Örneğime dayanarak, premium çekirdeklerin DAHA YÜKSEK puanlara yol açtığına dair güçlü kanıtlar var. Ekstra maliyete değerler!"</li>
    <li>H₀'yi reddetmeyi başaramadıysanız: "Örneğime dayanarak, premium çekirdeklerin daha iyi olduğuna dair yeterince güçlü bir kanıtım yok. Normal çekirdeklerle devam edip paradan tasarruf edeceğim."</li>
</ul>
